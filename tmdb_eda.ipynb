# TMDB EDA Assignment

This project fetches movie data from the TMDB API and performs exploratory data analysis (EDA) using Python and pandas.

The data is stored in a SQLite database and analyzed to gain insights.

## API Setup

In this step, we load the API key securely and import the required libraries.

```python
# NOTE: Replace with your own API key OR use a .env file

from dotenv import load_dotenv
import os
import requests
import pandas as pd
import sqlite3

load_dotenv()
API_KEY = os.getenv("API_KEY")
```

## Fetch Data from TMDB API

We use the TMDB Discover Movies endpoint to fetch data for at least 20 movies.

```python
import requests

url = "https://api.themoviedb.org/3/discover/movie"

params = {
    "api_key": API_KEY,   
    "language": "en-US",
    "page": 1
}

response = requests.get(url, params=params)

data = response.json()

print(data)
```

```python
movies = data['results']
```

## Convert Data to DataFrame

The JSON response from the API is converted into a pandas DataFrame for easier analysis.

```python
df = pd.DataFrame(movies)
df.head()
```

## Save Data to SQLite Database

The DataFrame is stored in a SQLite database table named 'movies' for persistence.

```python
# convert list column to string
df['genre_ids'] = df['genre_ids'].astype(str)

import sqlite3

conn = sqlite3.connect("movies.db")

df.to_sql("movies", conn, if_exists="replace", index=False)

conn.close()
```

## Load Data from SQLite

The stored data is loaded back into a pandas DataFrame for analysis.

```python
conn = sqlite3.connect("movies.db")

df = pd.read_sql("SELECT * FROM movies", conn)
```
